So, there are these things called "variables" I'm talking about. What even is a variable? Well, a variable is just a thing that I'm calling another thing! For example, we all know that 3+2=5!

So I could do something like this, where I define a variable called "x", set it equal to 3, and then add that to 2!

```python
x = 3
x + 2
```

OR, I could call 3 "thiscoolnumber", add it to two, and it will still equal 5!

```python
thiscoolnumber = 3
thiscoolnumber + 2
```
